```python
import pandas as pd
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import time

# Função para calcular o erro de previsão para um modelo SOM
def calculate_error(df, som, clusters):
    neuron_winners = np.array([som.winner(x) for x in data_values_scaled])
    df['neuron_winner_x'] = neuron_winners[:, 0]
    df['neuron_winner_y'] = neuron_winners[:, 1]

    # Calcular a precisão dos clusters para target = 0 e target = 1
    total_samples_per_cluster = [df[df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
    correct_samples_target_0_per_cluster = [df[df['target'] == 0][df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
    correct_samples_target_1_per_cluster = [df[df['target'] == 1][df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]

    precision_target_0_per_cluster = [correct / total if total > 0 else 0 for correct, total in zip(correct_samples_target_0_per_cluster, total_samples_per_cluster)]
    precision_target_1_per_cluster = [correct / total if total > 0 else 0 for correct, total in zip(correct_samples_target_1_per_cluster, total_samples_per_cluster)]

    # Calcular o erro médio quadrático para cada target
    mse_target_0 = mean_squared_error(total_samples_per_cluster, correct_samples_target_0_per_cluster)
    mse_target_1 = mean_squared_error(total_samples_per_cluster, correct_samples_target_1_per_cluster)

    return mse_target_0 + mse_target_1

# Carregar o arquivo CSV
file_path = 'nbr_with_lat_long.csv'
df = pd.read_csv(file_path)

# Ajustar os parâmetros
som_size = 4
som_portion = 0.5
sigma_values = [0.5, 1.0, 5.0, 10.0]
learning_rate_values = [0.0001, 0.001, 0.01, 0.1]
epocas = 100

# Selecionar as colunas de datas para a clusterização e as coordenadas
date_columns = df.columns[4:int(4 + 33/2)]  # Assumindo que as colunas de datas começam na 5ª coluna
data_values = df[date_columns].values
coordinates = df[['long', 'lat']].values

# Normalizar os dados temporais
scaler = StandardScaler()
data_values_scaled = scaler.fit_transform(data_values)
print(data_values_scaled)

# Loop para otimização
best_error = float('inf')
best_params = None
best_som = None

for sigma in sigma_values:
    for learning_rate in learning_rate_values:
        # Configuração e treinamento do SOM
        som = MiniSom(int(som_size * som_portion), som_size, data_values_scaled.shape[1], sigma, learning_rate)
        som.random_weights_init(data_values_scaled)
        
        # Listas para armazenar os erros durante o treinamento
        q_error = []
        t_error = []

        # Tempo inicial para o treinamento
        start_time = time.time()

        for i in range(epocas * len(data_values)):
            rand_i = np.random.randint(len(data_values))
            som.update(data_values_scaled[rand_i], som.winner(data_values_scaled[rand_i]), i, epocas)
            q_error.append(som.quantization_error(data_values_scaled))
            t_error.append(som.topographic_error(data_values_scaled))

        # Tempo final para o treinamento
        end_time = time.time()
        print(f"Tempo de execução (inicialização e treinamento): {end_time - start_time:.2f} segundos")

        # Plotar os erros
        plt.figure(figsize=(12, 6))
        plt.plot(np.arange(epocas * len(data_values)), q_error, label='Erro Quantização')
        plt.plot(np.arange(epocas * len(data_values)), t_error, label='Erro Topográfico')
        plt.ylabel('Erro')
        plt.xlabel('Iteração')
        plt.legend()
        plt.title(f'Erros durante o treinamento: sigma={sigma}, learning_rate={learning_rate}')
        plt.savefig(f'erros_treinamento_sigma_{sigma}_lr_{learning_rate}.png')
        plt.show()

        # Calcular o erro
        clusters = np.unique(np.array([som.winner(x) for x in data_values_scaled])[:, 0] * som_size + np.array([som.winner(x) for x in data_values_scaled])[:, 1])
        error = calculate_error(df, som, clusters)
        
        # Atualizar o melhor modelo
        if error < best_error:
            best_error = error
            best_params = (sigma, learning_rate)
            best_som = som

print(f"Melhores parâmetros encontrados: sigma={best_params[0]}, learning_rate={best_params[1]}")
print(f"Erro mínimo alcançado: {best_error}")

# Usar o melhor SOM para criar visualizações e tabelas
neuron_winners = np.array([best_som.winner(x) for x in data_values_scaled])
df['neuron_winner_x'] = neuron_winners[:, 0]
df['neuron_winner_y'] = neuron_winners[:, 1]

# Visualização dos clusters no mapa
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['long'], df['lat'], c=df['neuron_winner_x'] * som_size + df['neuron_winner_y'], cmap='viridis', marker='o')
plt.colorbar(scatter, label='Índice do Neurônio Vencedor')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição dos Dados no SOM')
plt.grid(True)
plt.savefig('distribuicao_dados_no_som.png')
plt.show()

# Plot da U-Matrix
plt.figure(figsize=(5, 5), dpi=100)
plt.imshow(best_som.distance_map(), cmap='Blues_r')
plt.title('U-Matrix')
plt.colorbar()
plt.savefig('u_matrix.png')
plt.show()

# Gráfico do target pelas coordenadas de latitude e longitude
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df['long'], df['lat'], c=df['target'], cmap='coolwarm', marker='o')
plt.colorbar(scatter, label='Target')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição do Target pelas Coordenadas')
plt.grid(True)
plt.savefig('distribuicao_target.png')
plt.show()

# Comparação dos valores dos clusters para target 0 e target 1
df_target_0 = df[df['target'] == 0]
df_target_1 = df[df['target'] == 1]

# Obter a lista de clusters
clusters = np.unique(df['neuron_winner_x'] * som_size + df['neuron_winner_y'])

# Calcular a precisão dos clusters para target = 0 e target = 1
total_samples_per_cluster = [df[df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
correct_samples_target_0_per_cluster = [df_target_0[df_target_0['neuron_winner_x'] * som_size + df_target_0['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
correct_samples_target_1_per_cluster = [df_target_1[df_target_1['neuron_winner_x'] * som_size + df_target_1['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]

precision_target_0_per_cluster = [correct / total if total > 0 else 0 for correct, total in zip(correct_samples_target_0_per_cluster, total_samples_per_cluster)]
precision_target_1_per_cluster = [correct / total if total > 0 else 0 for correct, total in zip(correct_samples_target_1_per_cluster, total_samples_per_cluster)]

# Criar DataFrames para as tabelas
cluster_results_target_0 = pd.DataFrame({
    'Cluster': clusters,
    'Total de Amostras': total_samples_per_cluster,
    'Amostras Corretas (Target=0)': correct_samples_target_0_per_cluster,
    'Precisão (Target=0)': precision_target_0_per_cluster
})

cluster_results_target_1 = pd.DataFrame({
    'Cluster': clusters,
    'Total de Amostras': total_samples_per_cluster,
    'Amostras Corretas (Target=1)': correct_samples_target_1_per_cluster,
    'Precisão (Target=1)': precision_target_1_per_cluster
})

# Salvar as tabelas em CSV
cluster_results_target_0.to_csv('cluster_results_target_0.csv', index=False)
cluster_results_target_1.to_csv('cluster_results_target_1.csv', index=False)

print("Tabela de Precisão para Target = 0 salva como 'cluster_results_target_0.csv'")
print("Tabela de Precisão para Target = 1 salva como 'cluster_results_target_1.csv'")

# Criar heat maps
heatmap_data_target_0 = np.array([correct_samples_target_0_per_cluster])
heatmap_data_target_1 = np.array([correct_samples_target_1_per_cluster])

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data_target_0, annot=True, cmap='Blues', xticklabels=[f'Cluster {i}' for i in clusters], yticklabels=['Target=0'])
plt.title('Heatmap de Amostras Corretas para Target = 0 por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Target')
plt.savefig('heatmap_target_0.png')
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data_target_1, annot=True, cmap='Reds', xticklabels=[f'Cluster {i}' for i in clusters], yticklabels=['Target=1'])
plt.title('Heatmap de Amostras Corretas para Target = 1 por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Target')
plt.savefig('heatmap_target_1.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from joblib import Parallel, delayed
from time import time

# Funções para calcular erros
def calculate_errors(som, data):
    # Calcular o erro de quantização
    quantization_error = np.mean([np.linalg.norm(x - som.weights[som.winner(x)]) for x in data])
    
    # Calcular o erro topográfico
    winners = np.array([som.winner(x) for x in data])
    topographic_errors = []
    for i, winner in enumerate(winners):
        # Verificar os vizinhos de um neurônio (8 vizinhos)
        neighbors = [(winner[0] + i, winner[1] + j) 
                     for i in [-1, 0, 1] 
                     for j in [-1, 0, 1] 
                     if (i, j) != (0, 0) and 0 <= winner[0] + i < som.get_weights().shape[0] and 0 <= winner[1] + j < som.get_weights().shape[1]]
        for neighbor in neighbors:
            if neighbor in winners:
                # Comparar se os neurônios vencedores são iguais
                if not np.any(np.all(winner == neighbors, axis=1)):
                    topographic_errors.append(1)
    topographic_error = np.mean(topographic_errors) if topographic_errors else 0

    return quantization_error, topographic_error

def run_som(sigma, learning_rate, epocas, data_values_pca):
    som_width = int(som_size * som_portion)
    som = MiniSom(som_width, som_size, data_values_pca.shape[1], sigma, learning_rate)

    # Inicialização dos pesos do SOM com os dois primeiros componentes principais do PCA
    weights_x, weights_y = np.meshgrid(
        np.linspace(data_values_pca[:, 0].min(), data_values_pca[:, 0].max(), som_width),
        np.linspace(data_values_pca[:, 1].min(), data_values_pca[:, 1].max(), som_size)
    )
    initial_weights = np.stack([weights_x.flatten(), weights_y.flatten()], axis=1).reshape(som_width, som_size, 2)
    som.weights = initial_weights

    # Armazenar erros ao longo das iterações
    q_error = []
    t_error = []

    # Treinamento do SOM e cálculo dos erros
    for i in range(epocas):
        som.train_batch(data_values_pca, 1)  # Treinar um batch
        q_error.append(np.mean([np.linalg.norm(x - som.weights[som.winner(x)]) for x in data_values_pca]))
        
        # Calcular o erro topográfico
        winners = np.array([som.winner(x) for x in data_values_pca])
        topographic_errors = []
        for j, winner in enumerate(winners):
            # Verificar os vizinhos de um neurônio (8 vizinhos)
            neighbors = [(winner[0] + i, winner[1] + j) 
                         for i in [-1, 0, 1] 
                         for j in [-1, 0, 1] 
                         if (i, j) != (0, 0) and 0 <= winner[0] + i < som.get_weights().shape[0] and 0 <= winner[1] + j < som.get_weights().shape[1]]
            for neighbor in neighbors:
                if neighbor in winners:
                    # Comparar se os neurônios vencedores são iguais
                    if not np.any(np.all(winner == neighbors, axis=1)):
                        topographic_errors.append(1)
        t_error.append(np.mean(topographic_errors) if topographic_errors else 0)

    return som, q_error, t_error

# Carregar o arquivo CSV
file_path = 'nbr_with_lat_long.csv'
df = pd.read_csv(file_path)

# Selecionar as colunas de datas para a clusterização e as coordenadas
date_columns = df.columns[20:22]  # Assumindo que as colunas de datas começam na 5ª coluna
data_values = df[date_columns].values
coordinates = df[['long', 'lat']].values

# Normalizar os dados temporais
scaler = StandardScaler()
data_values_scaled = scaler.fit_transform(data_values)

# Reduzir a dimensionalidade com PCA (2 componentes principais)
pca = PCA(n_components=2)
data_values_pca = pca.fit_transform(data_values_scaled)

# Parâmetros ajustáveis
som_size = 4
som_portion = 0.5
param_grid = {
    'sigma': [1.0, 5.0,10],
    'learning_rate': [0.001,0.01, 0.1],
    'epocas': [100, 500,1000]
}

# Buscar a melhor configuração
best_result = None
best_q_error = float('inf')
best_t_error = float('inf')

for sigma in param_grid['sigma']:
    for learning_rate in param_grid['learning_rate']:
        for epocas in param_grid['epocas']:
            som, q_error, t_error = run_som(sigma, learning_rate, epocas, data_values_pca)
            final_q_error = q_error[-1]
            final_t_error = t_error[-1]
            
            if final_q_error < best_q_error and final_t_error < best_t_error:
                best_q_error = final_q_error
                best_t_error = final_t_error
                best_result = {
                    'sigma': sigma,
                    'learning_rate': learning_rate,
                    'epocas': epocas,
                    'som': som,
                    'q_error': q_error,
                    't_error': t_error
                }

# Imprimir as melhores métricas
print(f"Melhor Configuração: {best_result}")
print(f"Erro de Quantização: {best_q_error}")
print(f"Erro Topográfico: {best_t_error}")

# Plotar os erros
plt.figure(figsize=(12, 6))

# Plotar o erro de quantização
plt.subplot(1, 2, 1)
plt.plot(best_result['q_error'], label='Erro de Quantização')
plt.xlabel('Iterações')
plt.ylabel('Erro de Quantização')
plt.title('Erro de Quantização ao Longo das Iterações')
plt.legend()
plt.grid(True)

# Plotar o erro topográfico
plt.subplot(1, 2, 2)
plt.plot(best_result['t_error'], label='Erro Topográfico', color='red')
plt.xlabel('Iterações')
plt.ylabel('Erro Topográfico')
plt.title('Erro Topográfico ao Longo das Iterações')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Carregar o arquivo CSV
file_path = 'nbr_with_lat_long.csv'
df = pd.read_csv(file_path)

# Ajustar os parâmetros
som_size = 4
som_portion = 0.8
sigma = 12.00
learning_rate = 0.001
epocas = 100

# Selecionar as colunas de datas para a clusterização e as coordenadas
date_columns = df.columns[19:21]  # Assumindo que as colunas de datas começam na 5ª coluna
# Extrair valores e normalizar
data_values = df[date_columns].values
data_values = data_values / np.max(np.sqrt(data_values * data_values))
coordinates = df[['long', 'lat']].values

# Normalizar os dados temporais
scaler = StandardScaler()
data_values_scaled = scaler.fit_transform(data_values)
print(data_values_scaled)

# Configuração e treinamento do SOM
som = MiniSom(int(som_size * som_portion), som_size, data_values_scaled.shape[1], sigma, learning_rate)
som.random_weights_init(data_values_scaled)
som.train_batch(data_values_scaled, epocas)  # Treinar por 1000 iterações

# Determinar os neurônios vencedores para cada ponto de dado
def get_winner_index(x):
    return som.winner(x)

neuron_winners = np.array([get_winner_index(x) for x in data_values_scaled])

# Mapear coordenadas para neurônios vencedores
df['neuron_winner_x'] = neuron_winners[:, 0]
df['neuron_winner_y'] = neuron_winners[:, 1]

# Visualização dos clusters no mapa
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['long'], df['lat'], c=df['neuron_winner_x'] * som_size + df['neuron_winner_y'], cmap='viridis', marker='o')
plt.colorbar(scatter, label='Índice do Neurônio Vencedor')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição dos Dados no SOM')
plt.grid(True)
plt.show()

# Plot da U-Matrix
plt.figure(figsize=(5, 5), dpi=100)
plt.imshow(som.distance_map(), cmap='Blues_r')
plt.title('U-Matrix')
plt.colorbar()
plt.show()

# Gráfico do target pelas coordenadas de latitude e longitude
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df['long'], df['lat'], c=df['target'], cmap='coolwarm', marker='o')
plt.colorbar(scatter, label='Target')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição do Target pelas Coordenadas')
plt.grid(True)
plt.show()

# Comparação dos valores dos clusters para target 0 e target 1
df_target_0 = df[df['target'] == 0]
df_target_1 = df[df['target'] == 1]

# Obter a lista de clusters
clusters = np.unique(df['neuron_winner_x'] * som_size + df['neuron_winner_y'])

# Calcular a média dos valores para cada cluster
mean_values_target_0 = [df_target_0[df_target_0['neuron_winner_x'] * som_size + df_target_0['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
mean_values_target_1 = [df_target_1[df_target_1['neuron_winner_x'] * som_size + df_target_1['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]

# Plotar a comparação
plt.figure(figsize=(12, 6))
bar_width = 0.35
clusters_indices = np.arange(len(clusters))

plt.bar(clusters_indices - bar_width / 2, mean_values_target_0, bar_width, label='Target = 0', color='blue')
plt.bar(clusters_indices + bar_width / 2, mean_values_target_1, bar_width, label='Target = 1', color='red')

plt.xlabel('Cluster')
plt.ylabel('Número de Amostras')
plt.title('Comparação dos Valores dos Clusters para Target 0 e Target 1')
plt.xticks(clusters_indices, [f'Cluster {i}' for i in clusters])
plt.legend()
plt.grid(True)
plt.show()

# Tabela de precisão dos clusters
total_samples_per_cluster = [df[df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster].shape[0] for cluster in clusters]
correct_samples_per_cluster = [df[df['neuron_winner_x'] * som_size + df['neuron_winner_y'] == cluster]['target'].sum() for cluster in clusters]
precision_per_cluster = [correct / total if total > 0 else 0 for correct, total in zip(correct_samples_per_cluster, total_samples_per_cluster)]

# Criar um DataFrame com os resultados
cluster_results = pd.DataFrame({
    'Cluster': clusters,
    'Total de Amostras': total_samples_per_cluster,
    'Amostras Corretas (Target=1)': correct_samples_per_cluster,
    'Precisão': precision_per_cluster
})

print(cluster_results)
mean_values_target_0
mean_values_target_1
um = [100*b/(a + b) for a, b in zip(mean_values_target_0, mean_values_target_1)]

zero = [100*a/(a + b) for a, b in zip(mean_values_target_0, mean_values_target_1)]
print(f'corretamente classificados como contendo queimada:{um} \n corretamente classificado como não contendo queimadas: {zero}')
print(f'Precisão média como contendo queimada: {sum(um)/int(som_size*som_size*0.8*0.5)} \nPrecisão média como não contendo queimada: {sum(zero)/int(som_size*som_size*0.8*0.5)}')
print(f'zero: {(mean_values_target_0)}')
print(f'Um {mean_values_target_1}')
```
